# 🌲 PrettyForest — Jupyter Demo

Interactive forest visualization for tree-based ML ensembles.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import polars as pl
from prettyforest import prettygrow
```

```python
# Load and split
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.3, random_state=42
)

# Train
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
model.fit(X_train, y_train)
print(f"Accuracy: {model.score(X_test, y_test):.1%}")
```

## Forest Visualization

The cell below renders the full interactive forest inline.
Try: hover trees, click to spotlight, trace a sample, double-click for detail view.

```python
X_test_pl = pl.DataFrame(
    {name: X_test[:, i] for i, name in enumerate(iris.feature_names)}
)
prettygrow(model, data=X_test_pl, target=y_test)
```

## Seasons

Pass  to set the initial theme (you can also switch live in the dropdown).

```python
prettygrow(model, data=X_test_pl, target=y_test)
```

## Boosted Model

Notice the different badges (gradient corrections) and the ⚠️ warning on double-click.

```python
from sklearn.ensemble import GradientBoostingClassifier

gbm = GradientBoostingClassifier(n_estimators=10, max_depth=3, random_state=42)
gbm.fit(X_train, y_train)
prettygrow(gbm, data=X_test_pl, target=y_test)
```
